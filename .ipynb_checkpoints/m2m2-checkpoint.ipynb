```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://www.redplanetscience.com'
```

```python
# Create BeautifulSoup object; parse with 'html.parser
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
list_elem = soup.select_one('div.list_text')

print(list_elem)
```

```python
# Retrieve the parent divs for all articles
results = soup.find_all('div', class_='list_text')

print(results)
```

```python
# loop over results to get article data
for result in results:
    # scrape the article title
    title = result.find('div', class_='content_title').get_text
    
    # scrape the article body
    body = result.find('div', class_='article_teaser_body').get_text    
    
    # print article data
    print('-----------------')
    print(title)
    print(body)
    
```
